```python
a=list(map(float,input("Enter the data under the attribte dollar mark-up on the statistics book:").split()))


def md(data):
  s=0
  for x in data:
    s+=x
  return(s/len(data))


def var(data,mean):
  ss=0
  for x in data:
    ss+=(x-mean)**2
  return (ss/len(data))


print('i.The mean dollar mark-up on the statistics book is =',md(a))
print('ii.The variance is =',round(var(a,md(a)),4))
```

```python
import math

def binomial_probability(n, k, p):

    binomial_coeff = math.comb(n, k)
    probability = binomial_coeff * (p * k) * ((1 - p) * (n - k))

    return probability

def average_users_watching_ad(n, p):
    return n * p


total_users = 200
probability_of_not_skip = 0.25

probability1 = binomial_probability(total_users, 40, probability_of_not_skip)
print(f"1. Probability that exactly 40 users will not skip the ad: {probability1:.4f}")

probability2 = binomial_probability(total_users, 60, probability_of_not_skip)
print(f"2. Probability that exactly 60 users will not skip the ad: {probability2:.4f}")

average_users = average_users_watching_ad(total_users, probability_of_not_skip)
print(f"3. Average number of users who will watch the ad completely: {average_users:.2f}")
```

```python
import random

def roll_dice():
  return random.randint(1, 9)

def get_probability(number):
  count = 0
  for i in range(100):
    value = roll_dice()
    if value == number:
      count += 1
  probability = count / 100
  return probability

def expected_value():
  sum_of_values = 0
  sum_of_probabilities = 0
  for i in range(1, 9):
    value = i
    probability = get_probability(i)
    sum_of_values += value * probability
    sum_of_probabilities += probability
  expected_value = sum_of_values / sum_of_probabilities
  return expected_value

def main():
  print("The probability of getting each number in the sample space is:")
  for i in range(1, 9):
    print(f"Number {i}: {get_probability(i)}")
  print("The expected value is:", expected_value())

if __name__ == "__main__":
  main()
```

```python
'''ravi invited ramya,who works at a call center,to dinner. ramya responded that she would join if
she isnt too tired after her work in the evening. while working, ramya in general, receives an
average of 2 calls per day.typically,she becomes tired if the number of calls she receive
in an 8-hour workday exceeds 20.
1)what is the probability of ramya receiving excatly 10 calls in a day?
2)what is the probability of ramya receiving excatly 15 calls in a day?
3)what is the probability of ravi and ramya not ahving dinner together?'''
import math
mu=2
def poisson_pmf(mu,x):
  return (math.e)*(-mu)*mu*x/math.factorial(x)
def poisson_cdf(mu,x):
  cdf=0
  for i in range(0,x):
    cdf=cdf+ poisson_pmf(mu,i)
  return cdf
print("(i):",poisson_pmf(mu,10))
print("(ii):",1-poisson_cdf(mu,15))
print("(iii):",1-poisson_cdf(mu,21))
#normal distribution
''' the annual salary of employees in a large company are approximately normally distributed with a mean of $50,000 and a standard deviation of $20,000.
i.what percent of employees earn less than $40,000?
ii.what percent of employees earn between $45,000 and $65,000?
iii.what percent of employees earn more than $70,000?'''
from scipy.stats import norm
mu=50000
sig=20000
def normal_cdf(mu,sig,x):
  z=(x-mu)/sig
  prob=norm.cdf(z)
  return prob
print("(i):",(normal_cdf(mu,sig,40000))*100)
print("(ii):",(normal_cdf(mu,sig,65000)-normal_cdf(mu,sig,45000))*100)
print(f'(iii):{(1 - normal_cdf(mu, sig, 70000)):.2f}')
```
